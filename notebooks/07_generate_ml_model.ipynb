```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import sys
from pathlib import Path

# Add the project root to the Python path
project_root = Path.cwd().parent
sys.path.append(str(project_root))

# Set up matplotlib for inline plotting
%matplotlib inline

def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path)
    df['Adjusted Date'] = pd.to_datetime(df['Adjusted Date'])
    df = df.sort_values('Adjusted Date')
    
    # Filter for specified date range and geounits
    df = df[(df['Adjusted Date'] >= '2023-01-01') & 
            (df['Adjusted Date'] <= '2024-06-30') &
            (df['Sl Geounit (Code)'].isin(['APG', 'QTG', 'ECP', 'NAO']))]
    
    categorical_features = ['Country Name', 'Billing Account', 'Rig Name', 'Well type']
    df[categorical_features] = df[categorical_features].fillna('Unknown')
    
    return df, categorical_features

file_path = project_root / "processed_data" / "processed_tickets_wles_ops_data.csv"
full_df, categorical_features = load_and_preprocess_data(file_path)

def train_and_predict(train_df, test_df, categorical_features):
    X_train = train_df[categorical_features + ['Unique_Well_Count']]
    y_train = train_df['Tickets_Revenue']
    X_test = test_df[categorical_features + ['Unique_Well_Count']]
    y_test = test_df['Tickets_Revenue']

    best_params = {
        'iterations': 383,
        'learning_rate': 0.05332604907347196,
        'depth': 5,
        'l2_leaf_reg': 2.523755213829783e-06,
        'bootstrap_type': 'Bernoulli',
        'subsample': 0.8737198107504663,
        'cat_features': categorical_features
    }

    model = CatBoostRegressor(**best_params)
    model.fit(X_train, y_train, eval_set=(X_test, y_test), early_stopping_rounds=30, verbose=100)

    test_preds = model.predict(X_test)

    return model, test_preds

def create_revenue_plot(geounit_df, geounit):
    geounit_df['Quarter'] = geounit_df['Adjusted Date'].dt.to_period('Q')
    
    train_df = geounit_df[geounit_df['Adjusted Date'] <= '2023-12-31']
    test_df = geounit_df[geounit_df['Adjusted Date'] >= '2024-01-01']
    
    model, test_preds = train_and_predict(train_df, test_df, categorical_features)
    
    test_df['Predicted'] = test_preds

    quarterly_full = geounit_df.groupby('Quarter').agg({
        'Tickets_Revenue': 'sum'
    }).reset_index()

    quarterly_test = test_df.groupby('Quarter').agg({
        'Tickets_Revenue': 'sum',
        'Predicted': 'sum'
    }).reset_index()

    quarterly_test['Percentage_Difference'] = ((quarterly_test['Predicted'] - quarterly_test['Tickets_Revenue']) / quarterly_test['Tickets_Revenue']) * 100

    fig, ax = plt.subplots(figsize=(15, 8))

    # Plot actual revenue for the entire timeline (2022-10-01 to 2024-06-30)
    ax.plot(quarterly_full['Quarter'].astype(str), quarterly_full['Tickets_Revenue'], label='Actual', marker='o', color='blue')

    # Plot predicted revenue for the test period
    ax.plot(quarterly_test['Quarter'].astype(str), quarterly_test['Predicted'], label='Predicted (Test)', marker='s', color='green', linestyle=':')

    ax.set_title(f'Actual vs Predicted Revenue by Quarter ({geounit} Geounit)')
    ax.set_xlabel('Quarter')
    ax.set_ylabel('Revenue')
    ax.legend()
    ax.tick_params(axis='x', rotation=45)

    # Add percentage difference annotations for test period
    for i, row in quarterly_test.iterrows():
        ax.annotate(f"{row['Percentage_Difference']:.2f}%", 
                    (str(row['Quarter']), row['Predicted']),
                    textcoords="offset points", 
                    xytext=(0,10), 
                    ha='center')

    plt.tight_layout()
    plt.show()

    print(f"Quarterly Test Data for {geounit}:")
    print(quarterly_test)

    return model

# Create plots for each geounit
geounits = ['APG', 'QTG', 'ECP', 'NAO']
models = {}

for geounit in geounits:
    geounit_df = full_df[full_df['Sl Geounit (Code)'] == geounit]
    models[geounit] = create_revenue_plot(geounit_df, geounit)

# Model Feature Importance for each geounit
for geounit, model in models.items():
    feature_importance = model.feature_importances_
    feature_names = categorical_features + ['Unique_Well_Count']

    feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': feature_importance})
    feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

    plt.figure(figsize=(10, 6))
    plt.bar(feature_importance_df['feature'], feature_importance_df['importance'])
    plt.title(f'Feature Importance ({geounit} Geounit)')
    plt.xlabel('Features')
    plt.ylabel('Importance')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print(f"\nFeature Importance for {geounit}:")
    print(feature_importance_df)
```
